```python
import torch
from transformers import SpeechT5ForTextToSpeech, SpeechT5Processor
import soundfile as sf

# Load the pretrained SpeechT5 model and processor
processor = SpeechT5Processor.from_pretrained("microsoft/speecht5_tts")
model = SpeechT5ForTextToSpeech.from_pretrained("microsoft/speecht5_tts")

# Define the input text
text = "This is a demonstration of text-to-speech using a pretrained model."

# Preprocess the input text
inputs = processor(text=text, return_tensors="pt")

# Generate speech
with torch.no_grad():
    speech = model.generate_speech(inputs["input_ids"], speaker_embeddings=None)

# Save the generated speech to a .wav file
sf.write("output.wav", speech.numpy(), 16000)
print("Speech synthesis complete. Saved to 'output.wav'.")
```
